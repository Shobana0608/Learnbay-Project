# Loading of Data set

```python
import pandas as pd
df=pd.read_excel(r'/content/CarMarketResearchDataset.xlsx')
print(df)
```

```python
df.columns = [c.replace(' ','_') for c in df.columns]
x_tr=df.Opinion
y_tr=df.Driver_Name
```

```python
df.columns
```

## Finding the number of clusters into which the opinions have been classified according to the dataset given to us

```python
count = 1
each = 'road presence'
old = 'road presence'
for each in y_tr:
  if each != old:
    count = count+1
    old = each
numberofclusters = count
print(numberofclusters)
```

# Getting 'just nouns' versions of all opinions - We will name it as 'NounArray'


```python
import nltk
nltk.download('punkt_tab')
nltk.download('averaged_perceptron_tagger_eng')
```

```python
from textblob import TextBlob
nounarray = []
for opinion in x_tr:
  concatenate = ""
  blob = TextBlob(opinion)
  txt = [n for n,t in blob.tags if t == 'NN']
  if len(txt) == 0:
    txt.append(opinion)
  for token in txt:
      concatenate = concatenate + " " + token
  nounarray.append(concatenate)
nounarray
```

## Getting feature vectors (Word Embeddings) of all contents of NounArray using TFIDF Vectorizer

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
import numpy as np
import pandas as pd
```

```python
vectorizer = TfidfVectorizer(stop_words = 'english')
X= vectorizer.fit_transform(nounarray)
```

```python
true_k=3
model = KMeans(n_clusters=true_k, init='k-means++', max_iter=100, n_init=1)
model.fit(X)
```

```python
model.predict(X)
```

# Getting centroids of each cluster in terms of a group of 'm' number of words (Here m = 10)

```python
order_centroids = model.cluster_centers_.argsort()[:, ::-1]
terms = vectorizer.get_feature_names_out()
```

```python
for i in range(true_k):
  print ("Cluster ", i, " \n"),
  for ind in order_centroids[i, :10]:
    print(terms[ind])
```

## Clustering and Printing the clusters

```python
prediction = []
for i in nounarray:
  X = vectorizer.transform([i])
  predicted = model.predict(X)
  prediction.append(predicted)
print (prediction)

```

```python
finalclusters = []
for i in range(true_k):
  finalclusters.append([])

for counter in range(len(prediction)):
  finalclusters[prediction[counter][0]].append(x_tr[counter])

for eachclusterindex in range(len(finalclusters)):
  print("Cluster ", eachclusterindex, " \n")
  print(finalclusters[eachclusterindex], "\n")
```

## calculating accuracies

```python
# for road presence
rp = []
for i in range(len(y_tr)):
    if y_tr[i] == 'road presence':
        rp.append(x_tr[i])

d = 0
calcrp = finalclusters[2]
print ("Supposedly misclassified sentences: \n")
for i in calcrp:
    if i in rp:
        d = d + 1
    else:
        print (i)
if d == len(calcrp):
    print ("None")
print("\n Accuracy: ", d/len(calcrp))
```

```python
# for Build Quality
bd = []
for i in range(len(y_tr)):
    if y_tr[i] == 'build quality':
        bd.append(x_tr[i])

d = 0
# Corrected: finalclusters[1] is for 'build quality'
calcbd = finalclusters[0]
print ("Supposedly misclassified sentences: \n")
for i in calcbd:
    if i in bd:
        d = d + 1
    else:
        print (i)
if d == len(calcbd):
    print ("None")
print("\n Accuracy: ", d/len(calcbd))
```

```python
#For Safety Features

sf = []
for i in range(len(y_tr)):
    if y_tr[i] == "safety features":
        sf.append(x_tr[i])

d = 0
calcsf = finalclusters[2]
print ("Supposedly misclassified sentences: \n")
for i in calcsf:
    if i in sf:
        d = d + 1
    else:
        print (i)
if d == len(calcsf):
    print ("None")
print("\n Accuracy: ",d/len(calcsf))
```
